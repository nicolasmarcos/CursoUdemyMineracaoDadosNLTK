```python
#stopwords nltk
stopwordsnltk = nltk.corpus.stopwords.words('portuguese')
print(stopwordsnltk)
```

```python
stopwords = ()
#print(stopwords)
#print(set(stopwordsnltk))
#print(set(stopwordsmanuais))

stopwords = set(stopwordsnltk).union(set(stopwordsmanuais))

print(stopwords)

```

```python
def removeStopWords(texto):
    frases = []
    #Para cada palavra e emoção da minha base de dados percorreremos texto
    for (palavras, emocao) in texto:
        # para cada palavra p splitada verá se consta no conjunto de stopwords. 
        # senão, aplica a variável semstop
        semstop = [p for p in palavras.split() if p not in stopwords ]
        frases.append((semstop,emocao))
        #print(frases)
    return frases
```

```python
print(removeStopWords(basetreinamento))
```

```python
def aplicaStemmer(texto):
    #Stemmer de PT
    stemmer = nltk.stem.RSLPStemmer()
    frasessteming = []
    for (palavras,emocao) in texto:
        # O for anterior passa uma frase;
        # O stemmer.stem(p) aplicará stemming a uma palavra P extraindo seu radical
        # para cada palavra do conjunto de subpalavras
        comstemming = [str(stemmer.stem(p)) for p in palavras.split() if p not in stopwords]
        frasessteming.append((comstemming, emocao))
    return frasessteming    
```

```python
frasesComStemmingTreinamento = aplicaStemmer(basetreinamento)
print(frasesComStemmingTreinamento)
```

```python
def buscapalavras(frases):
    todaspalavras = []
    for (palavras,emocao) in frases:
        todaspalavras.extend(palavras)
    return todaspalavras
```

```python
palavrasTreinamento = buscapalavras(frasesComStemmingTreinamento)
```

```python
print(palavrasTreinamento)
```

```python
def buscafrequencia(palavras):
    palavras = nltk.FreqDist(palavras)
    return palavras
```

```python
frequenciaTreinamento = buscafrequencia(palavrasTreinamento)
```

```python
print(frequenciaTreinamento.most_common(10))
```

```python
def buscapalavrasunicas(frequencia):
    freq = frequencia.keys()
    return freq
```

```python
palavrasUnicasTreinamento = buscapalavrasunicas(frequenciaTreinamento)
```

```python
print(palavrasUnicasTreinamento)
```

```python
# A ideia é percorrer um documento e verificar se existe dentre as palavras únicas

def extratorpalavras(documento):
    doc = set(documento)
    caracteristicas = {}
    for palavras in palavrasUnicasTreinamento:
        caracteristicas['%s' % palavras] = (palavras in doc)
    return caracteristicas
```

# Passo 2 - Treinamento Modelo

```python
baseCompletaTreinamento = nltk.classify.apply_features(extratorpalavras, frasesComStemmingTreinamento)
```

```python
# Constrói a tabela de probabilidade

classificador = nltk.NaiveBayesClassifier.train(baseCompletaTreinamento)

# Lista classes / rótulos identificados
print(classificador.labels())
```
